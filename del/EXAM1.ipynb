```python
    # import statements

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

```

```python
# load data

df = pd.read_excel('Indonesia_airport.xlsx')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# Checking which columns have null values
df.isnull().sum()
```

```python
df['Island'].value_counts()
# replace null values with most reoccuring value
print(df['Island'].mode()[0])
# most reoccuring value is Jawa
df['Island'].fillna(df['Island'].mode()[0], inplace=True)
df['Time Zone'].fillna(df['Time Zone'].mode()[0], inplace=True)
df['Special Status'].fillna(df['Special Status'].mode()[0], inplace=True)

# Applying mean to numerical values to replace null values
df['Total Urban Villages'].fillna(df['Total Urban Villages'].mean(), inplace=True)
df['Total Rural Villages'].fillna(df['Total Rural Villages'].mean(), inplace=True)
df['Growth Factor of New Cases'].fillna(df['Growth Factor of New Cases'].mean(), inplace=True)
df['Growth Factor of New Deaths'].fillna(df['Growth Factor of New Deaths'].mean(), inplace=True)

df.head()
```

```python
df.isnull().sum()

# All Null values have been replaced
```

```python
# Numerical EDA

# Mono-variate Analysis
# Histograms
df.hist(figsize=(20, 10))
plt.show()

```

```python
# Box plot of each column
```

```python
# Bar graph of each column

plt.figure(figsize=(20, 10))
plt.bar(df.columns, df.mean())
plt.show()
```

```python
# Bi-variate analysis between two columns with relationship

# Scatter plot
plt.figure(figsize=(20, 10))
plt.scatter(df['Total Urban Villages'], df['Total Rural Villages'])
plt.show()

```

```python
# Convert date time
df['Date'] = pd.to_datetime(df['Date'])

# Convert date to columns such as day, month, year
df['Year'] = pd.to_datetime(df['Date']).dt.year
df['Month'] = pd.to_datetime(df['Date']).dt.month
df['Day'] = pd.to_datetime(df['Date']).dt.day

# Convert location ISO to numerical
df['Location ISO Code'] = pd.Categorical(df['Location ISO Code'])
df['Location ISO Code'] = df['Location ISO Code'].cat.codes



```

```python
# Correlation matrix between numerical columns

plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), annot=True)
plt.show()

```

```python
# Categorical EDA 

# Bar graph of each column
plt.figure(figsize=(20, 10))
plt.bar(df['Island'].value_counts().index, df['Island'].value_counts())
plt.show()




```

```python
# Frequency Distribution of all columns

for col in df.columns:
    df[col].value_counts().plot(kind='bar')
    plt.show()
    
    


```

```python
# One hot encoding

df = pd.get_dummies(df, columns=['Island', 'Time Zone', 'Special Status'], drop_first=True)
df.head()
```

```python
# Line Graph
plt.figure(figsize=(20, 10))
plt.plot(df['Date'], df['Total Cases per Million'])
plt.plot(df['Date'], df['New Cases per Million'])

# convert values to date time


plt.show()
```

```python

# Pie chart
plt.figure(figsize=(20, 10))
plt.pie(df['Island'].value_counts(), labels=df['Island'].value_counts().index, autopct='%1.1f%%')
```
